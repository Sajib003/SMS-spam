```python
import pandas as pd
import re
import string
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.preprocessing import LabelEncoder
```

```python
pd.set_option('display.max_colwidth',100)
df=pd.read_csv('spam_msg.csv', encoding='latin-1')
```

```python
print(df.info())
print(df.head())
```

```python
df = df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4','Unnamed: 5'],axis=1)
print(df)
df.groupby('v1').describe()
```

```python
df = df.rename(columns = {'v1':'label','v2':'message'})
df.head()
```

```python
X=LabelEncoder()
df.iloc[:,0]=X.fit_transform(df.iloc[:,0])
df.head()
```

```python
df['length'] = df['message'].apply(len)
def punc_count(txt):
    count=sum([1 for i in txt if i in string.punctuation])
    return 100*(count/len(txt))
df['punctuation']=df['message'].apply(lambda x: punc_count(x))
df.to_csv (r'C:\Users\Sajib\Desktop\SMS\new_data.csv', index = None, header=True)
```

```python
ps = PorterStemmer()
corpus=[]
tok=[]
tok2=[]
```

```python
for i in range(0, 5755):
    msg = df['message'][i]
    msg = re.sub('^\w+@[a-zA-Z_]+?\.[a-zA-Z]{2,3}$', 'emailaddr', df['message'][i])
    msg = re.sub('(http[s]?\S+)|(\w+\.[A-Za-z]{2,4}\S*)', 'httpaddr', df['message'][i])
    msg = re.sub('£|\$', 'moneysymb', df['message'][i])
    msg = re.sub('((\(\d{3}\) ?)|(\d{3}-))?\d{3}-\d{4}', 'phonenumbr', df['message'][i])
    msg = re.sub('\d+(\.\d+)?', 'numbr', df['message'][i])
    msg = re.sub('[!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~]','',msg)
    
    if i==0:
        print("\t\t\t\t MESSAGE ", i)
    if i==0:
        print("\n After Regular Expression - Message ", i, " : ", msg)
        
    msg = msg.lower()    
    if i==0:
        print("\n Lower case Message ", i, " : ", msg)
    
    # Tokenize
    msg = msg.split()    
    if i==0:
        print("\n After Splitting - Message ", i, " : ", msg)
    
    # Stemming with Stop Words
    msg2 = [ps.stem(word) for word in msg if not word in set(stopwords.words('english'))]
    tok.append(msg2)
    if i==0:
        print("\n After Stemming - Message ", i, " : ", msg2)
    
    # preparing Messages with Remaining Tokens
    msg4 = ' '.join(msg2)
    if i==0:
        print("\n Final Prepared - Message ", i, " : ", msg4, "\n\n")

    # Preparing WordVector Corpus
    corpus.append(msg4)
    
df['message'] = corpus
df['token'] = tok
df.to_csv (r'C:\Users\Sajib\Desktop\SMS\new_data2.csv', index = None, header=True)
```
